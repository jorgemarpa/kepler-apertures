```python
import os, glob
import pickle
import numpy as np
import pandas as pd

from astropy.io import fits
from astropy.table import Table
```

```python
channels = np.delete(np.arange(1, 85), [4,5,6,7])
quarters = np.arange(5, 18)
```

```python
rknots = {}
phiknots = {}
df_dict = {}

for q in quarters:
    quarter_table = []
    for ch in channels:
        fname = "../data/models/%i/channel_%02i_psf_model.pkl" % (
                    q,
                    ch,
                )
        if os.path.isfile(fname):
            psf = pickle.load(open(fname, "rb"))
        
        rknots["q%i_ch%i" % (q, ch)]   = psf["n_r_knots"]
        phiknots["q%i_ch%i" % (q, ch)] = psf["n_phi_knots"]
        
        quarter_table.append([psf["n_r_knots"], psf["n_phi_knots"], psf["rmin"], psf["rmax"]] + psf["psf_w"].tolist())
    aux = pd.DataFrame(np.array(quarter_table), index=channels,
                       columns=["n_r_knots", "n_phi_knots", "rmin", "rmax"] + ["b%02i" % i for i in range(1, 1 + len(psf["psf_w"]))],
                      )
    df_dict[q] = aux.astype({"n_r_knots": int, "n_phi_knots": int})
```

```python
primary_hdu = fits.PrimaryHDU(data=None)
phdr = primary_hdu.header
phdr.set("OBJECT", "PRF models")
phdr.set("EXTNAME", "PRIMARY")
```

```python
table_hdus = []
for q in df_dict.keys():
    cols = []
    for col in df_dict[q].columns:
        column = fits.Column(name="ch%02i" % int(df_dict[q].loc["channel", col]), format="D", array=df_dict[q].loc[:,col][1:])
        cols.append(column)
    coldef = fits.ColDefs(cols)
    hdu = fits.BinTableHDU.from_columns(coldef)
    hdu.header["EXTNAME"] = "QUARTER_%02i" % q
    hdu.header["QUARTER"] = q
    hdu.header["ROW0"] = "n_r_knots"
    hdu.header["ROW1"] = "n_phi_knots"
    hdu.header["ROW2"] = "rmin"
    hdu.header["ROW3"] = "rmax"
    hdu.header["ROW4"] = "prf_w"
    table_hdus.append(hdu)
```

```python
hdul = fits.HDUList([primary_hdu] + table_hdus)
```
